```python
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1

df_mod =df[~ ((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis =1)]
df_mod.head()
```

#### Now move to select the feature that are highly correlated with class categorigal varible



```python
for i in range (1, 27):
    col = f'V{i}'
    correlation = df[col].corr(df['Class'])
    print("The correlation b/w",f'V{i}',  "and class is :", correlation)
```

```python
from scipy.stats import ttest_ind, mannwhitneyu

# Separate fraud and non-fraud
fraud = df_mod[df_mod['Class'] == 1]
non_fraud = df_mod[df_mod['Class'] == 0]

# Features to test  choose ones with high correlation
features_to_test = ['Amount'] + [f'V{i}' for i in range(1,27)]

print("---- Hypothesis Testing Results ----\n")

for col in features_to_test:
    # Use Welchs t-test (doesn't assume equal variance)
    t_stat, p_val = ttest_ind(fraud[col], non_fraud[col], equal_var=False)

    # Optional: Use Mann-Whitney U as non-parametric backup
    u_stat, u_p_val = mannwhitneyu(fraud[col], non_fraud[col])

    print(f"Feature: {col}")
    print(f"  T-test p-value: {p_val:.5f}")
    print(f"  Mann-Whitney U p-value: {u_p_val:.5f}")
    
    if p_val < 0.05:
        print("   Statistically significant difference\n")
    else:
        print("   No significant difference\n")
```

```python
from sklearn.feature_selection import f_classif

X = df_mod.drop('Class', axis=1)
y = df_mod['Class']

f_values, p_values = f_classif(X, y)

anova_result = pd.DataFrame({
    'Feature': X.columns,
    'F-value': f_values,
    'p-value': p_values
})
anova_result['p-value'] = anova_result['p-value'].apply(lambda x: f"{x:.30f}")
print(anova_result)
```
